# Scatter Plot
A scatter plot is used to visualize the relationship between two continuous variables

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt


ds = pd.read_csv('/content/athlete_events.csv')
#print(ds.head(5))
sns.scatterplot(x='Weight',y='Height', hue='Sex', size='Age', data =ds)
plt.title("Relationship between Weight and Height")
plt.show()
```

#Bar plot
Bar plots are used to visualize the relationship between a categorical variable and a continuous variable

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt


ds = pd.read_excel('/content/Costofanightout.xlsx')
#print(ds.head(5))
sns.barplot(x='Category',y='Cost',data=ds, hue='Country', palette="muted", errorbar=None )
plt.title("Cost of Nightout across World")
plt.show()
```
