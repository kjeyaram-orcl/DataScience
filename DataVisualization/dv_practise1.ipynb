# Scatter Plot
A scatter plot is used to visualize the relationship between two continuous variables

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt


ds = pd.read_csv('/content/athlete_events.csv')
#print(ds.head(5))
sns.scatterplot(x='Weight',y='Height', hue='Sex', size='Age', data =ds)
plt.title("Relationship between Weight and Height")
plt.show()
```

#Bar plot
Bar plots are used to visualize the relationship between a categorical variable and a continuous variable

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt


ds = pd.read_excel('/content/Costofanightout.xlsx')
#print(ds.head(5))
sns.barplot(x='Category',y='Cost',data=ds, hue='Country', palette="muted", errorbar=None )
plt.title("Cost of Nightout across World")
plt.show()
```

```python
#groupby and aggregation example

import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


ds = pd.read_csv('/content/ur.csv')
#print(ds.head(5))

res= ds.groupby('Country Name').aggregate({'Unemployement Rate':np.mean})
print(res)

sns.barplot(x='Country Name',y='Unemployement Rate',data=res )
plt.show()
```

#Line plots
Line plots are used to visualize trends in data over time or other continuous variables.

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt


ds = pd.read_csv('/content/ur.csv')
#print(ds.head(5))

sns.lineplot(x='Date',y='Unemployement Rate',hue='Country Name', data=ds)
plt.title("Unemployment Rate by Country")
plt.show()


sns.lineplot(x='Date',y='Unemployement Rate',hue='Country Name', style='Gender', data=ds)
plt.title("Unemployment Rate by Country and Gender")
plt.show()
```

#Histograms
Histograms visualize the distribution of a continuous variable. In a histogram, the data is divided into bins and the height of each bin represents the frequency or count of data points within that bin

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt


ds = pd.read_excel('/content/Costofanightout.xlsx')
#print(ds.head(5))

sns.histplot(x='Cost', data=ds, bins=10, kde=True , color='green')
plt.title("Distribution of Cost in Nightout ")
plt.show()
```

#Density plots
Density plots also known as kernel density plots, are a type of data visualization that display the distribution of a continuous variable.

```python

import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt


ds = pd.read_excel('/content/Costofanightout.xlsx')
#print(ds.head(5))
sns.kdeplot(data=ds, x="Cost", hue="Category", fill=True)
plt.title("Distribution of Cost in Nightout by Category ")
plt.show()
```

#Box plots
Box plots are a type of visualization that shows the distribution of a dataset. They are commonly used to compare the distribution of one or more variables across different categories.

```python
ds = pd.read_excel('/content/Costofanightout.xlsx')
#print(ds.head(5))
sns.boxplot(data=ds, x="Category", y="Cost", hue='Item', palette="Set3")
plt.title("Box Plot of Cost by Day and Item ")
plt.show()
```

#Violin plots
A violin plot is a type of data visualization that combines aspects of both box plots and density plots

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt


ds = pd.read_csv('/content/ur.csv')
#print(ds.head(5))

sns.violinplot(x='Country Name',y='Unemployement Rate', data=ds)
plt.title("Unemployment Rate by Country")
plt.show()
```

#Heatmaps
A heatmap is a graphical representation of data that uses colors to depict the value of a variable in a two-dimensional space.

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt


ds = pd.read_csv('/content/athlete_events.csv')
#print(ds.head(5))
sns.heatmap(ds.corr(),annot=True)
plt.show()
```

```python
import seaborn as sns

import matplotlib.pyplot as plt

flights = sns.load_dataset('flights')
flights = flights.pivot('month', 'year', 'passengers')
sns.heatmap(flights, cmap='Blues', annot=True, fmt='d')
plt.title('Passengers per month')
plt.xlabel('Year')
plt.ylabel('Month')
plt.show()
```

#pair plots
Pair plots are a type of visualization in which multiple pairwise scatter plots are displayed in a matrix format.

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt


ds = pd.read_csv('/content/athlete_events.csv')
#print(ds.head(5))
sns.pairplot(data=ds)
plt.show()
```

#Joint plot
Joint plot is a powerful visualization technique in seaborn that combines two different plots in one visualization: a scatter plot and a histogram.

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt


ds = pd.read_csv('/content/athlete_events.csv')
sns.jointplot(x="Weight", y="Height", data=ds)
plt.show()
```
