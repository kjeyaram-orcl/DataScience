```python
import pandas as pd
technologies = {
    'Courses':["Spark","PySpark","Python","pandas"],
    'Fee' :[20000,25000,22000,30000],
    'Duration':['30days','40days','35days','50days'],
              }
index_labels=['r1','r2','r3','r4']
df1 = pd.DataFrame(technologies,index=index_labels)

#print("First DataFrame:\n", df1)

technologies2 = {
    'Courses':["Spark","Java","Python","Go"],
    'Discount':[2000,2300,1200,2000]
              }
index_labels2=['r1','r6','r3','r5']
df2 = pd.DataFrame(technologies2,index=index_labels2)

#print("Second DataFRame:\n", df2)

df3=df1.join(df2, lsuffix="_left", rsuffix="_right", how='inner')
print("After inner joining two DataFrames:\n", df3)

df3=df1.join(df2, lsuffix="_left", rsuffix="_right", how='left')
print("After left joining two DataFrames:\n", df3)

df3=df1.join(df2, lsuffix="_left", rsuffix="_right", how='right')
print("After right joining two DataFrames:\n", df3)

df3=df1.join(df2, lsuffix="_left", rsuffix="_right", how='outer')
print("After outer joining two DataFrames:\n", df3)


df3=df1.set_index('Courses').join(df2.set_index('Courses'), how='inner') # Join on Column
print("After Join on Column:\n",df3)

df3=df1.join(df2.set_index('Courses'), how='inner', on='Courses')# Join on Column
print("After Join on Column:\n",df3)
```

```python
import pandas as pd
technologies = {
    'Courses':["Spark","PySpark","Python","pandas"],
    'Fee' :[20000,25000,22000,30000],
    'Duration':['30days','40days','35days','50days'],
              }
index_labels=['r1','r2','r3','r4']
df1 = pd.DataFrame(technologies,index=index_labels)

technologies2 = {
    'Courses':["Spark","Java","Python","Go"],
    'Discount':[2000,2300,1200,2000]
              }
index_labels2=['r1','r6','r3','r5']
df2 = pd.DataFrame(technologies2,index=index_labels2)

# Merge by column
df3=pd.merge(df1,df2, on='Courses')
print(df3)

# Merge on different colunn names
df3=pd.merge(df1,df2, left_on='Courses', right_on='Courses')
print(df3)

# Merge by Index
df3 = pd.merge(df1,df2,left_index=True,right_index=True)
print(df3)
```

```python
import pandas as pd
df1 = pd.DataFrame({'Courses': ["Spark","PySpark","Python","pandas"],
                    'Fee' : [20000,25000,22000,24000]})

df2 = pd.DataFrame({'Courses': ["Pandas","Hadoop","Hyperion","Java"],
                    'Fee': [25000,25200,24500,24900]})

df3 = pd.concat([df1, df2], ignore_index=True, sort=False)
print("After concatenation:\n",df3)

df3 = pd.concat([df1, df2], axis=1, join='inner')
print("After column concatenation:\n",df3)
```
