```python
#Simple linear regression is a statistical method you can use to understand the relationship between two variables, x and y.


import matplotlib.pyplot as plt
from scipy import stats

x = [2,4,6,8]
y = [3,7,5,10]

slope, intercept, r, p, std_err = stats.linregress(x, y)

def myfunc(x):
  return slope * x + intercept

mymodel = list(map(myfunc, x))

plt.scatter(x, y)
plt.plot(x, mymodel)
plt.show()
```

```python
#Simple linear regression is a statistical method you can use to understand the relationship between two variables, weight and height.


import matplotlib.pyplot as plt
from scipy import stats

x = [140,155,159,179,192,200,212]
y = [60,62,67,70,71,72,75]

slope, intercept, r, p, std_err = stats.linregress(x, y)

def myfunc(x):
  return slope * x + intercept

mymodel = list(map(myfunc, x))

plt.scatter(x, y)
plt.plot(x, mymodel)
plt.show()

# The coefficient of determination can range from 0 to 1.
#A value of 0 indicates that the response variable cannot be explained by the predictor variable at all.
#A value of 1 indicates that the response variable can be perfectly explained without error by the predictor variable
print("coefficient of determination (R-squared) is ", r**2, ", This tells us that weight is a very good predictor of height.")
```
