# Scatter Plot
A scatter plot is used to visualize the relationship between two continuous variables

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt


ds = pd.read_csv('/content/athlete_events.csv')
#print(ds.head(5))
sns.scatterplot(x='Weight',y='Height', hue='Sex', size='Age', data =ds)
plt.title("Relationship between Weight and Height")
plt.show()
```

#Bar plot
Bar plots are used to visualize the relationship between a categorical variable and a continuous variable

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt


ds = pd.read_excel('/content/Costofanightout.xlsx')
#print(ds.head(5))
sns.barplot(x='Category',y='Cost',data=ds, hue='Country', palette="muted", errorbar=None )
plt.title("Cost of Nightout across World")
plt.show()
```

#Line plots
Line plots are used to visualize trends in data over time or other continuous variables.

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt


ds = pd.read_csv('/content/ur.csv')
#print(ds.head(5))

sns.lineplot(x='Date',y='Unemployement Rate',hue='Country Name', data=ds)
plt.title("Unemployment Rate by Country")
plt.show()


sns.lineplot(x='Date',y='Unemployement Rate',hue='Country Name', style='Gender', data=ds)
plt.title("Unemployment Rate by Country and Gender")
plt.show()
```

#Histograms
Histograms visualize the distribution of a continuous variable. In a histogram, the data is divided into bins and the height of each bin represents the frequency or count of data points within that bin

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt


ds = pd.read_excel('/content/Costofanightout.xlsx')
#print(ds.head(5))

sns.histplot(x='Cost', data=ds, bins=10, kde=True , color='green')
plt.title("Distribution of Cost in Nightout ")
plt.show()
```

```python

import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt


ds = pd.read_excel('/content/Costofanightout.xlsx')
#print(ds.head(5))
sns.kdeplot(data=ds, x="Cost", hue="Category")
plt.title("Distribution of Cost in Nightout by Category ")
plt.show()
```
